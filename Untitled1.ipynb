```python
import bs4
```

```python
import requests
```

```python
res=requests.get(
    "https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%EB%A7%9B%EC%A7%91")
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
ul_elements=soup.select("ul")
```

```python
type(ul_elements) #파이썬 기본자료형인 list
```

```python
ul_elements[0]
```

```python
type(ul_elements[0])
```

```python
ul_elements[1]
```

```python
ul_elements=soup.select("ul.lst_total")
ul_elements
len(ul_elements)
```

```python
# 해당 ul의 정보를 가져오는 부분
ul_element=ul_elements[0]
ul_element
```

```python
li_elements=ul_element.select("li")
```

```python
len(li_elements) #네이버 뷰의 맛집이 30개 list가 있음
```

```python
li_elements=soup.select("ul.lst_total li")
len(li_elements)
```

```python
# li의 _svp_item의 포함된 자료를 가져온다. 
li_elements=soup.select("ul.lst_total li._svp_item")
len(li_elements)
```

```python
# li의 _svp_item의 직계 자식들만 가져온다. 
li_elements=soup.select("ul.lst_total > li._svp_item")
len(li_elements)
```
